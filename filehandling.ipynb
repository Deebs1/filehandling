```python
weather_dict = {}
with open("nyc_weather.csv") as f:
    for line in f:
        tokens = line.split(',')
        # print(tokens)
        day = tokens[0]
        # print(day)
        try:
            temperature = int(tokens[1])
            # print(temperature)
            weather_dict[day] = temperature
            # print(weather_dict)
        except:
            print("Invalid temperature, ignore the row")   
```

```python
word_count = {}
with open("poem.txt","r") as f:
    for line in f:
        tokens = line.split(' ')
        # print(tokens)
        for token in tokens:
            token=token.replace('\n','')
            # print(token)
            if token in word_count:
                word_count[token]+=1
                # print(word_count)
            else:
                word_count[token]=1
                # print(word_count)
```
